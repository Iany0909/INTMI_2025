```python
opcao = input('Digite "S" para continuar ou "N" para sair: ').strip().upper()

if opcao == "S":
    print("Você escolheu continuar.")
elif opcao == "N":
    print("Você escolheu sair.")
else:
    print("Opção inválida.")
```

```python
dias = ["Domingo", "Segunda-feira", "Terça-feira", "Quarta-feira",
        "Quinta-feira", "Sexta-feira", "Sábado"]

numero = int(input("Digite um número de 1 a 7: "))

if 1 <= numero <= 7:
    print(dias[numero - 1])
else:
    print("Número inválido! Digite um número entre 1 e 7.")
```

```python
# Programa para verificar se um ano é bissexto
ano = int(input("Digite um ano: "))
# Verifica se o ano é bissexto
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"{ano} é um ano bissexto.")
else:
    print(f"{ano} não é um ano bissexto.")
```

```python

# Solicita os dados ao usuário
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_hora = float(input("Digite o valor recebido por hora: "))

# Verifica se há horas extras
if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_hora
else:
    horas_extras = horas_trabalhadas - 40
    salario = (40 * valor_hora) + (horas_extras * valor_hora * 1.5)

# Exibe o resultado
print(f"Seu salário total é: R$ {salario:.2f}")
```
